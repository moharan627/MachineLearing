データセットの準備（ボストンの住宅価格）

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['target'] = np.array(boston.target)
data = df.drop(['target'],axis=1)
target = df['target']
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=2022)
```

行列Aを取得。推論。

```python
X = X_train.values
y = y_train.values
A = np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)
y_train_pre = np.dot(X,A)

Xt = X_test.values
y_pre = np.dot(Xt,A)
```

学習用データと試験用データで誤差取得。

```python
from sklearn.metrics import mean_squared_error
print("train:",mean_squared_error(y_train, y_train_pre))
print("test:",mean_squared_error(y_test, y_pre))
```
